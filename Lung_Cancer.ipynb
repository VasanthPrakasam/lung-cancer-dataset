```python
from sqlalchemy import create_engine

# Create a SQLAlchemy engine with connection pooling and recycling
# Replace {username}, {password}, {host}, {port}, and {database} with your actual database credentials
engine = create_engine(
  "mysql+mysqlconnector://{username}:{password}@{host}:{port}/{database}",
    pool_size=10,  # Increased pool size for better performance with multiple connections
    pool_recycle=3600,  # Recycle connections after 1 hour to prevent stale connections
    pool_pre_ping=True  # Test connections before use to ensure they are active
)
```

```python
# Load the cleaned data from the CSV to ensure we are using the latest version
cleaned_df = pd.read_csv('cleaned_lung_cancer_data.csv', index_col='id')

# Define chunk size for inserting data into the database
chunk_size = 5000

# Calculate the number of chunks based on the DataFrame size and chunk size
num_chunks = len(cleaned_df) // chunk_size + (len(cleaned_df) % chunk_size != 0)

print(f"Loading data in {num_chunks} chunks...")

# Load data in chunks into the 'cleaned_lung_cancer_data' table in the database
for i in range(num_chunks):
    start_row = i * chunk_size
    end_row = (i + 1) * chunk_size
    df_chunk = cleaned_df[start_row:end_row]

    if i == 0:
        # For the first chunk, replace the table if it exists before inserting
        df_chunk.to_sql("cleaned_lung_cancer_data", con=engine, if_exists='replace', index=True)
        print(f"Chunk {i+1}/{num_chunks} loaded (replaced table).")
    else:
        # For subsequent chunks, append to the existing table
        df_chunk.to_sql("cleaned_lung_cancer_data", con=engine, if_exists='append', index=True)
        print(f"Chunk {i+1}/{num_chunks} loaded (appended).")

print("Data loading complete.")
```

```python
# Load the raw data from the original CSV file
raw_df = pd.read_csv('/content/Lung Cancer.csv')

# Define chunk size for inserting raw data into the database
chunk_size = 5000

# Calculate the number of chunks for the raw data
num_chunks = len(raw_df) // chunk_size + (len(raw_df) % chunk_size != 0)

print(f"Loading raw data into 'mytable' in {num_chunks} chunks...")

# Load raw data in chunks into the 'mytable' in the database
for i in range(num_chunks):
    start_row = i * chunk_size
    end_row = (i + 1) * chunk_size
    df_chunk = raw_df[start_row:end_row]

    if i == 0:
        # For the first chunk of raw data, replace the table if it exists
        df_chunk.to_sql("mytable", con=engine, if_exists='replace', index=False) # Set index=False as 'id' is not the index in the raw data
        print(f"Chunk {i+1}/{num_chunks} loaded (replaced table).")
    else:
        # For subsequent chunks of raw data, append to the existing table
        df_chunk.to_sql("mytable", con=engine, if_exists='append', index=False) # Set index=False
        print(f"Chunk {i+1}/{num_chunks} loaded (appended).")

print("Raw data loading complete.")
```

```python
# Define a SQL query to select the first 10 rows from 'mytable'
query = "SELECT * FROM mytable LIMIT 10" # Limiting to 10 rows for a quick view
# Read the data from the database using the SQL query into a DataFrame
raw_data_from_sql = pd.read_sql(query, con=engine)
# Display the DataFrame containing the raw data from the database
display(raw_data_from_sql)
```
